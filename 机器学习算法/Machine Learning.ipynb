```python
import tushare as ts
from pprint import pprint
import time
import pandas as pd
import numpy
from numpy import nan
```

```python
# 目录
# 一、读取原始数据，获得基本面DataFrame，各股报表发布日期DataFrame
# 二、计算Y（是否大涨）-读取各股一段时间内的收盘价格，获得股票与收盘价序列的Dictionary
# 三、计算Y（是否大涨）-通过各股发布日期与测试日期对应的股价计算Y
# 四、处理基本面DataFrame获得X
# 五、X变量字典
# 六、获得训练集的X与Y并保存输出
# 七、仿照一至六获得验证集的X与Y并保存输出
```

```python
# 一、读取原始数据

# 1. 读取股票基本面数据函数
def basic_inf_stock(year,quarter):
    
    """
    参数说明
    year      : int 报表所属年份
    quarter   : int 1~4，报表所属季度
    """
    
    df_profit = ts.get_profit_data(year,quarter)
    df_operation = ts.get_operation_data(year,quarter)
    df_growth = ts.get_growth_data(year,quarter)
    df_debtpaying = ts.get_debtpaying_data(year,quarter)
    df_cashflow = ts.get_cashflow_data(year,quarter)
    return df_profit,df_operation,df_growth,df_debtpaying,df_cashflow

# 2. 读取股票报表报告日期与n月后日期函数
def report_inf_stock(year,quarter,n):
    
    """
    参数说明
    year      : int 报表所属年份
    quarter   : int 1~4，报表所属季度
    n         : int 测试日期与发布日期的间隔月数
    """
        
    # 读取
    df_report = ts.get_report_data(year,quarter)
    
    # 构造股票代码、报告时期、测试时期的处理后DataFrame，股票代码是索引
    year_str = str(year) + '-'
    df_report_time = pd.DataFrame()
    df_report_time['report_date'] = df_report['report_date'].apply(lambda x: year_str + x)
    M_second = n*30*24*60*60
    df_test_time = pd.DataFrame()
    df_test_time['test_date'] = df_report_time['report_date'].apply(lambda x: time.strftime("%Y-%m-%d",time.localtime((time.mktime(time.strptime(x,'%Y-%m-%d')) + M_second))))
    df_temp_0 = pd.concat([df_report['code'],df_report_time,df_test_time],axis=1)
    
    # 对df_dd_rt_tt进行索引去重
    df_dd_rt_tt = (df_temp_0.drop_duplicates(keep='first')).set_index('code')
    
    return df_dd_rt_tt,df_report

# 3. 运行函数
df_profit,df_operation,df_growth,df_debtpaying,df_cashflow = basic_inf_stock(2016,4)
df_report_date,df_report_inf = report_inf_stock(2016,4,7/30)
```

```python
# 二、读取各股票发布日与一月后收盘价格

# 1. 定义批量读取函数
def stock_inf_get(L,start_date,end_date):
    
    """
    参数说明
    L         : list 股票代码的列表，例如['000002','000423',...]
    start_date: str  数据起始时间，格式YYYY-MM-DD
    end_date  : str  数据结束时间，格式YYYY-MM-DD
    """
    tol = len(L)
    cnt = 0
    dict_stock_all = {}
    for stock_code in L:
        
        # 记录抓取进度
        cnt = cnt + 1
        
        # 以防未抓取到数据报错
        df_test = ts.get_k_data(stock_code,start = start_date,end = end_date)
        if len(df_test) == 0:
            continue
        
        # 处理抓取到的数据
        df_temp = df_test[['date','close']]
        df = df_temp.set_index('date')
        dict_stock_all[stock_code] = df
        print(cnt,'/',tol)
        
    return dict_stock_all

def run_all(L,start_date,end_date):
    
    """
    参数说明
    L         : list 股票代码的列表，例如['000002','000423',...]
    start_date: str  数据起始时间，格式YYYY-MM-DD
    end_date  : str  数据结束时间，格式YYYY-MM-DD
    outputpath: str  输出路径，例如'c:/XXX/YYYYY'
    """
    df_dict = stock_inf_get(L,start_date,end_date)
    return df_dict

# 2. 设定样本股票代码列表
# df_zz500_index_stock = ts.get_zz500s()
train_stock_list = list(df_report_date.index)

# 3. 运行函数获得列表内股票的价格信息
all_stock_dict = run_all(L = train_stock_list,start_date = '2016-01-01',end_date = '2016-07-01')
```

```python
# 三、通过各股发布日期与测试日期对应的股价计算Y，即各个股票在公布年报后的一个月后是否涨幅超过阈值

# 1. 定义函数实现
def find_if_up(L, df_rt_tt, dict_code_price, threshold):
    
    """
    参数说明
    L               : list 股票代码的列表，例如['000002','000423',...]
    df_rt_tt        : DataFrame 包含年报发布时间report_date与测试涨幅时间test_date两列与code索引列
    dict_code_price : dict 键-值对为股票代码-DataFrame对，DataFrame包含close一列与date索引列
    threshold       : float 判定股票为涨的阈值
    """
    # 通过迭代股票代码列表中的每一支股票，获取股票报表发布日期与测试日期，获取对应日期下的股票价格，计算涨跌幅
    
    df_save = pd.DataFrame(columns = ['code','Y'])
    day_1 = 1*24*60*60
    cnt = 0
    for stock_code in L:
        
        # 跳过所找股票不在公布年报的股票代码列表内的情况
        if (stock_code not in list(df_rt_tt.index)) or (stock_code not in list(dict_code_price.keys())):
            continue

        # 获得该股票对应的年报公布时间与一个月后的测试时间
        rept_date = df_rt_tt['report_date'][stock_code]
        test_date = df_rt_tt['test_date'][stock_code]
        
        # 确保该股票年报公布时间与一个月后测试时间是交易日，跳过太多（停牌等）也删去
        check1 = 0
        while rept_date not in list(dict_code_price[stock_code].index) and check1 < 10:
            rept_date = time.strftime("%Y-%m-%d",time.localtime((time.mktime(time.strptime(rept_date,'%Y-%m-%d')) + day_1)))
            check1 = check1 + 1
        if check1 == 10:
            continue
        check2 = 0
        while test_date not in list(dict_code_price[stock_code].index) and check2 < 10:
            test_date = time.strftime("%Y-%m-%d",time.localtime((time.mktime(time.strptime(test_date,'%Y-%m-%d')) + day_1))) 
            check2 = check2 + 1
        if check2 == 10:
            continue
        
        # 读取年报公布时间与一个月后测试时间的股价并计算收益率
        p_r = dict_code_price[stock_code]['close'][rept_date]
        p_t = dict_code_price[stock_code]['close'][test_date]
        growth = (p_t - p_r)/p_r
        
        if growth > threshold:
            y = 1
        else:
            y = 0
        df_save.loc[cnt] = [stock_code, y]
        cnt = cnt + 1
    return df_save

# 2. 运行函数获得DataFrame，['code','growth','Y']
df_Y_train = find_if_up(L = train_stock_list, df_rt_tt = df_report_date, dict_code_price = all_stock_dict, threshold = 0.05)
# df_Y_train.to_excel('C:/Users/43460/Desktop/y.xlsx',sheet_name = 'y_train', index=False)
```

```python
# 四、处理基本面DataFrame获得X

# 输入六个DataFrame返回数值化、去重处理后的一个DataFrame即为X
def get_X_DataFrame(d1,d2,d3,d4,d5,d6):
    
    """
    参数说明
    d1 :df_report_inf
    d2 :df_profit
    d3 :df_operation
    d4 :df_growth
    d5 :df_debtpaying
    d6 :df_cashflow
    """
    
    # 所有DataFrame去重
    d1 = d1.drop_duplicates(keep='first')
    d2 = d2.drop_duplicates(keep='first')
    d3 = d3.drop_duplicates(keep='first')
    d4 = d4.drop_duplicates(keep='first')
    d5 = d5.drop_duplicates(keep='first')
    d6 = d6.drop_duplicates(keep='first')

    # 构建临时DataFrame去除股票中文名等无用变量
    df_temp_a = d1[['code','eps','eps_yoy','roe','net_profits','profits_yoy','distrib']]
    df_temp_b = d2[['code','roe','net_profit_ratio','gross_profit_rate','net_profits','eps','business_income','bips']]
    df_temp_c = d3[['code','arturnover','arturndays','inventory_turnover','inventory_days','currentasset_turnover','currentasset_days']]
    df_temp_d = d4[['code','mbrg','nprg','nav','targ','epsg','seg']]
    df_temp_e = d5[['code','currentratio','quickratio','cashratio','icratio','sheqratio','adratio']]
    df_temp_f = d6[['code','cf_sales','rateofreturn','cf_nm','cf_liabilities','cashflowratio']]

    # 合并为一张X的宽表df_X_train
    df_temp = pd.merge(left = df_temp_a, right = df_temp_b, how = 'inner', left_on = 'code', right_on = 'code')
    df_temp = pd.merge(left = df_temp, right = df_temp_c, how = 'inner', left_on = 'code', right_on = 'code')
    df_temp = pd.merge(left = df_temp, right = df_temp_d, how = 'inner', left_on = 'code', right_on = 'code')
    df_temp = pd.merge(left = df_temp, right = df_temp_e, how = 'inner', left_on = 'code', right_on = 'code')
    df_X_train = pd.merge(left = df_temp, right = df_temp_f, how = 'inner', left_on = 'code', right_on = 'code')
    df_X_train['distrib'] = df_X_train['distrib'].apply(lambda x: 0 if type(x) is float else 1)
    X_list = list(df_X_train.columns)
    X_list.remove('code')
    for lsst in X_list:
        df_X_train[lsst] = df_X_train[lsst].apply(lambda x: 0 if x == '--' else float(x))
        
    return df_X_train

# 运行函数获得X
df_X_train = get_X_DataFrame(df_report_inf,df_profit,df_operation,df_growth,df_debtpaying,df_cashflow)
```

```python
# 五、X变量字典

# esp,每股收益
# eps_yoy,每股收益同比(%)
# bvps,每股净资产
# roe,净资产收益率(%)
# epcf,每股现金流量(元)
# net_profits,净利润(万元)
# profits_yoy,净利润同比(%)
# distrib,分配方案,1为有分配,0为无分配

# roe,净资产收益率(%)
# net_profit_ratio,净利率(%)
# gross_profit_rate,毛利率(%)
# net_profits,净利润(万元)
# eps,每股收益
# business_income,营业收入(百万元)
# bips,每股主营业务收入(元)

# arturnover,应收账款周转率(次)
# arturndays,应收账款周转天数(天)
# inventory_turnover,存货周转率(次)
# inventory_days,存货周转天数(天)
# currentasset_turnover,流动资产周转率(次)
# currentasset_days,流动资产周转天数(天)

# mbrg,主营业务收入增长率(%)
# nprg,净利润增长率(%)
# nav,净资产增长率
# targ,总资产增长率
# epsg,每股收益增长率
# seg,股东权益增长率

# currentratio,流动比率
# quickratio,速动比率
# cashratio,现金比率
# icratio,利息支付倍数
# sheqratio,股东权益比率
# adratio,股东权益增长率

# cf_sales,经营现金净流量对销售收入比率
# rateofreturn,资产的经营现金流量回报率
# cf_nm,经营现金净流量与净利润的比率
# cf_liabilities,经营现金净流量对负债比率
# cashflowratio,现金流量比率
```
