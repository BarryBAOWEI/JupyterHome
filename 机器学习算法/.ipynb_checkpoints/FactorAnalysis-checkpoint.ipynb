```python
import pandas as pd
from factor_analyzer import FactorAnalyzer
from sklearn.preprocessing import scale
import numpy as np
```

```python
df = pd.read_excel('C:/Users/jxjsj/Desktop/stat.xlsx')
# 标准化
df = df.apply(lambda x: (x-x.mean())/x.std())
```

```python
factorA = FactorAnalyzer(n_factors=3, rotation='varimax')
result = factorA.fit(df)
```

```python
# 原始数据中变量的相关系数矩阵
result.corr_
```

```python
# 因子载荷矩阵
result.loadings_
```

```python
# 每个变量的公共度：因子系数的平方和
result.get_communalities()
```

```python
# 特征根
result.get_eigenvalues()
```

```python
# 第三行，因子对应主成分的累积方差贡献度
result.get_factor_variance()
```

```python
# 因子得分
result.transform(df)
```

```python
# 因子得分系数矩阵
import statsmodels.api as sm
factor_score_list = []
for factor in range(result.transform(df).shape[1]):
    est = sm.OLS(result.transform(df)[:,factor],df)
    result_ = est.fit()
    factor_score_list.append(list(result_.params))
factor_score_matrix = np.array(factor_score_list).T
factor_score_matrix
```
