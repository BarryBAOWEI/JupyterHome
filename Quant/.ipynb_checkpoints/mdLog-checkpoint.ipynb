```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_0 = pd.read_csv('C:/Users/jxjsj/Desktop/mdLog.csv')
```

```python
### 变量构造 ###

# 买卖价差
df_0['bid-ask spread'] = df_0['ask']-df_0['bid']

# 市场规模
df_0['market size'] = (df_0['bidsize'] + df_0['asksize'])/2

# 中间价
df_0['mid'] = (df_0['ask']+df_0['bid'])/2

# 差分
df_0['deltaBAS'] = df_0['bid-ask spread'] - df_0['bid-ask spread'].shift(1)
df_0['deltaBid'] = df_0['bid']- df_0['bid'].shift(1) 
df_0['deltaAsk'] = df_0['ask'] - df_0['ask'].shift(1)
df_0['deltaMid'] = df_0['mid'] - df_0['mid'].shift(1)
```

```python
# How many times did the bid-ask spread widen ?
df_BAspreadPos = df_0[df_0['deltaBAS']>0]
len_BAspreadPos = len(df_BAspreadPos)
print(len_BAspreadPos)
```

```python
# What proportion of those times did it widen on both sides (bid decreased and ask increased)?
df_BAspreadPos1 = df_BAspreadPos[(df_BAspreadPos.deltaBid<0) & (df_BAspreadPos.deltaAsk>0)]
both_side = len(df_BAspreadPos1)
print(both_side/len_BAspreadPos*100 ,'%')
```

```python
# What proportion of those times did it widen on one side (bid decreased xor ask increased)?
df_BAspreadPos2 = df_BAspreadPos[(df_BAspreadPos.deltaBid<0) | (df_BAspreadPos.deltaAsk>0)]
one_side = len(df_BAspreadPos2) - both_side
print(one_side/len_BAspreadPos*100 ,'%')
```

```python
# How many times did the bid-ask spread tighten?
df_BAspreadNeg = df_0[df_0.deltaBAS<0]
len_BAspreadNeg = len(df_BAspreadNeg)
print(len_BAspreadNeg)
```

```python
print(df_0['market size'].describe())
dataForDis = df_0['market size']
plt.hist(dataForDis, 100)

plt.show()
```

```python
# Report the distribution of the market sizes of the defensive side 
# and aggressive side immediately following a change in mid price. 
# Can you intuitively explain your results?

df_MidChangeBASNot = df_0[(df_0['deltaMid'] != 0) & (df_0['deltaBAS'] == 0)]
# Aggressive
df_Agg_1 = df_MidChangeBASNot[df_MidChangeBASNot['deltaMid']>0][['bidsize']]
df_Agg_1.columns=['Agg']
df_Agg_2 = df_MidChangeBASNot[df_MidChangeBASNot['deltaMid']<0][['asksize']]
df_Agg_2.columns=['Agg']
df_Agg = pd.concat([df_Agg_1,df_Agg_2],axis=0)
plt.subplot(1,2,1)
plt.hist(df_Agg['Agg'])
# Defensive
df_Def_1 = df_MidChangeBASNot[df_MidChangeBASNot['deltaMid']>0][['asksize']]
df_Def_1.columns=['Def']
df_Def_2 = df_MidChangeBASNot[df_MidChangeBASNot['deltaMid']<0][['bidsize']]
df_Def_2.columns=['Def']
df_Def = pd.concat([df_Def_1,df_Def_2],axis=0)
plt.subplot(1,2,2)
plt.hist(df_Def['Def'])

print(df_Agg.describe())
print(df_Def.describe())

plt.tight_layout()
plt.show()

print('进攻方防御方出现的次数是相同的，实际上的规律很可能是相近的相等是个巧合；防御方的挂单数量的量级明显高于进攻方，这是因为在中间价出现变动时，防御方总是有利的一方，会促进防御方进行交易')
```
