```python
from scipy.optimize import fsolve
import math
from scipy.integrate import quad
import numpy as np
import datetime
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3
```

```python
'''
辅助函数，服从标准正态分布随机变量pdf与cdf
'''
```

```python
def dN(x):
    '''pdf'''
    return math.exp(-0.5 * x ** 2) / math.sqrt(2 * math.pi)

def N(x):
    '''cdf'''
    return quad(lambda x: dN(x), -np.inf, x)[0]

def d1f(St, K, t, T, r, sigma):
    ''' Black-Scholes-Merton d1 function.
        Parameters see e.g. BSM_call_value function. '''
    Tmt = (T - t).days / 365.
    d1 = (math.log(St / K) + (r + 0.5 * sigma ** 2)
          * (Tmt)) / (sigma * math.sqrt(Tmt))
    return d1
```

```python
'''
BSM期权定价公式
'''
```

```python
def BSM_call_value(St, K, t, T, r, sigma):

    d1 = d1f(St, K, t, T, r, sigma)
    Tmt = (T - t).days / 365.
    d2 = d1 - sigma * math.sqrt(Tmt)
    call_value = St * N(d1) - math.exp(-r * (Tmt)) * K * N(d2)
    return call_value


def BSM_put_value(St, K, t, T, r, sigma):

    Tmt = (T - t).days / 365.
    put_value = BSM_call_value(St, K, t, T, r, sigma) \
        - St + math.exp(-r * (Tmt)) * K
    return put_value
```

```python
'''
隐含波动率求解
'''
```

```python
def implied_vol(St, K, t, T, r, Vt, start_sigma=0.2, callOrPut = 0):
    '''
    St, 标的当前价
    K, 期权行权价
    t, 当前日期
    T, 合约到期日期
    r, 连续无风险利率
    Vt, 期权价格
    start_sigma, 隐含波动估计初始值 
    callOrPut, 0为call 1为put
    '''
    if callOrPut == 0:
        BSM_value =  BSM_call_value
    if callOrPut == 1:
        BSM_value =  BSM_put_value
    def f(x):
        return np.array(BSM_value(St,K,t,T,r,x)-Vt)
    iv = fsolve(f, [start_sigma])[0]
    return iv
```

```python
'''
例子1 - 给定2019-06-12日 50ETF 6月2450认购期权 价值0.3569以及50ETF现货价值2.802，计算隐含波动率
例子2 - 假设价值未知需要计算，新加入历史现货年化波动率（240日数据计算）0.2406
'''
```

```python
t = datetime.datetime.strptime('2019-06-12','%Y-%m-%d')
T = datetime.datetime.strptime('2019-06-26','%Y-%m-%d')
St = 2.802
K = 2.45
r_ = 0.02439 # 到期时间跨度相近的shibor，2W
r = 26*np.log(1+r_/26)
Vt = 0.3569
sigma_est = 0.2406

print('1.ImpliedVol',implied_vol(St, K, t, T, r, Vt, start_sigma=0.2, callOrPut = 0))
print('2.BSMOptionPrice',BSM_call_value(St, K, t, T, r,  sigma_est),'realPrice',Vt)
```

```python
'''
希腊字母
'''
```

```python
def eur_option_greeks(St, K, t, T, r, sigma_est, Tmt=None, callOrPut = 0):
#     ImpliedVol = implied_vol(St, K, t, T, r, Vt, start_sigma=0.2, callOrPut = callOrPut)
    if Tmt:
        pass
    else:
        Tmt = (T - t).days / 365.
    d1 = d1f(St, K, t, T, r, sigma_est)
    d2 = d1 - sigma_est * np.sqrt(Tmt)
    N_d1 = N(d1)
    N_d2 = N(d2)
    dN_d1 = dN(d1)
    if callOrPut == 0:
        delta = N_d1
        gamma = dN_d1/(St*sigma_est*np.sqrt(Tmt))
        theta = -(St*dN_d1*sigma_est)/(2*np.sqrt(Tmt))-r*np.exp(-r*Tmt)*K*N_d2
        rho = K*Tmt*np.exp(-r*Tmt)*N_d2
        vega = St*dN_d1*np.sqrt(Tmt)
    if callOrPut == 1:
        delta = 1 - N_d1
        gamma = dN_d1/(St*sigma_est*np.sqrt(Tmt))
        theta = -(St*dN_d1*sigma_est)/(2*np.sqrt(Tmt))+r*np.exp(-r*Tmt)*K*N(-d2)
        rho = -K*Tmt*np.exp(-r*Tmt)*N(-d2)
        vega = St*dN_d1*np.sqrt(Tmt)
    return {'delta':delta,'gamma':gamma,'theta':theta,'rho':rho,'vega':vega}
```

```python
'''
例子1 - 给定2019-06-12日 50ETF 6月2450认购期权 价值0.3569以及50ETF现货价值2.802，计算Greeks
'''
```

```python
t = datetime.datetime.strptime('2019-06-12','%Y-%m-%d')
T = datetime.datetime.strptime('2019-06-26','%Y-%m-%d')
St = 2.802
K = 2.45
r_ = 0.02439 # 到期时间跨度相近的shibor，2W
r = 26*np.log(1+r_/26)
Vt = 0.3569
sigma_est = 0.2406

eur_option_greeks(St, K, t, T, r, sigma_est, callOrPut = 0)
```

```python
'''
希腊字母三维图 - (到期时间,行权价格,Greek)
'''
```

```python
def plot_greeks(St, K, t, T, r, sigma_est, callOrPut = 0):
    
    # X轴执行价格以及Y轴到期时间
    tlist = np.linspace(0.01, 1, 25)
    klist = np.linspace(K*0.8, K*1.2, 25)
    
    for greek in ['delta','gamma','theta','rho','vega']:    
        V = np.zeros((len(tlist), len(klist)), dtype=np.float)
        for j in range(len(klist)):
            for i in range(len(tlist)):
                # Z轴希腊字母值，此处传入了Tmt，实际上t与T就不发挥作用了
                V[i, j] = eur_option_greeks(St, klist[j], t, T, r, sigma_est, Tmt = tlist[i], callOrPut = callOrPut)[greek]

        # 3D Plotting
        x, y = np.meshgrid(klist, tlist)
        fig = plt.figure(figsize=(9, 5))
        plot = p3.Axes3D(fig)
        plot.plot_wireframe(x, y, V)
        plot.set_xlabel('strike $K$')
        plot.set_ylabel('maturity $T$')
        plot.set_zlabel('%s(K, T)' % greek)
```

```python
plot_greeks(St, K, t, T, r, sigma_est, callOrPut = 0)
```
