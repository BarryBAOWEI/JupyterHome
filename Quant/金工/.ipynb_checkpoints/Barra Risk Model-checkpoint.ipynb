```python
def applyParallel(dfGrouped, func):
    retLst = Parallel(n_jobs=multiprocessing.cpu_count())(delayed(func)(group) for name, group in dfGrouped)
    return pd.concat(retLst)

def compute_beta(df):
    df = df.copy()
    code = df['code'].tolist()[0]
    # rolling
    wind_len = 252
    df_ = df[['day','return','markbench_return']]
    save_dict = {}
    for start in range(len(df_)-wind_len):
        end = start + wind_len
        df_sub = df_.iloc[start:end,:]
        date = df_sub['day'].tolist()[-1]
        try:
            X = df_sub['return'].dropna()
            y = df_sub['markbench_return'].dropna()
            X = sm.add_constant(X)
            beta = sm.OLS(y,X).fit().params[1]
            save_dict[date] = beta
        except:
            save_dict[date] = np.nan
    save_df = pd.DataFrame.from_dict(save_dict,orient='index').reset_index().rename(columns={'index':'day',0:'beta'})
    save_df['code'] = code
    return save_df
```

```python
beta_tmp2 = applyParallel(beta_tmp1.groupby('code'),compute_beta)
```

```python
def applyParallel(dfGrouped, func):
    retLst = Parallel(n_jobs=multiprocessing.cpu_count())(delayed(func)(group) for name, group in dfGrouped)
    return pd.concat(retLst)

def compute_ewm_beta(df):
    df = df.copy()
    df['Beta'] = df['beta'].ewm(halflife=63).mean()
    return df
```

```python
def ewa(series,halflife):
    x = []
    alpha = 1 - np.exp(np.log(0.5) / halflife)
    x.append(series[0])
    for i in range(1,len(series)):
        x.append((1-alpha)*x[i-1]+alpha*series[i])
    return x[-1]
```

```python
def compute_dastd(df):
    df = df.copy()
    df['rolling252_mean'] = df['return'].rolling(252).mean()
    
    code = df['code'].tolist()[0]
    # rolling
    wind_len = 252
    df_ = df[['day','return','rolling252_mean']]
    save_dict = {}
    for start in range(len(df_)-wind_len):
        end = start + wind_len
        df_sub = df_.iloc[start:end,:].copy()
        date = df_sub['day'].tolist()[-1]
        return_mean = df_sub['rolling252_mean'].tolist()[-1]
        try:
            df_sub['return_'] = df_sub['return'].apply(lambda x: (x-return_mean)**2)
            dastd = ewa(df_sub['return_'].tolist(),42)    
            save_dict[date] = dastd
        except:
            save_dict[date] = np.nan
    save_df = pd.DataFrame.from_dict(save_dict,orient='index').reset_index().rename(columns={'index':'day',0:'dastd'})
    save_df['code'] = code
    
    return save_df
```

```python
def applyParallel(dfGrouped, func):
    retLst = Parallel(n_jobs=multiprocessing.cpu_count())(delayed(func)(group) for name, group in dfGrouped)
    return pd.concat(retLst)

dastd_tmp0 = applyParallel(stock_trade_data_tmp5.groupby('code'), compute_dastd)
```

```python
def compute_hsigma(df):
    df = df.copy()
    code = df['code'].tolist()[0]
    # rolling
    wind_len = 252
    df_ = df[['day','return','markbench_return']]
    save_dict = {}
    for start in range(len(df_)-wind_len):
        end = start + wind_len
        df_sub = df_.iloc[start:end,:]
        date = df_sub['day'].tolist()[-1]
        try: 
            X = df_sub['return'].dropna()
            y = df_sub['markbench_return'].dropna()
            X = sm.add_constant(X)
            res = sm.OLS(y,X).fit().resid
            hsigma = np.std(res)
            save_dict[date] = hsigma
        except:
            save_dict[date] = np.nan
    save_df = pd.DataFrame.from_dict(save_dict,orient='index').reset_index().rename(columns={'index':'day',0:'hsigma'})
    save_df['code'] = code
    return save_df
```

```python
def applyParallel(dfGrouped, func):
    retLst = Parallel(n_jobs=multiprocessing.cpu_count())(delayed(func)(group) for name, group in dfGrouped)
    return pd.concat(retLst,sort=False)

hsigma_tmp0 = applyParallel(hsigma_tmp1.groupby('code'), compute_hsigma)
```

```python
def compute_crma(df):
    df = df.copy()
    code = df['code'].tolist()[0]
    # rolling
    wind_len = 252
    df_ = df[['day','return']]
    save_dict = {}
    for start in range(len(df_)-wind_len):
        end = start + wind_len
        df_sub = df_.iloc[start:end,:].copy()
        date = df_sub['day'].tolist()[-1]
        try:
            df_sub['return_'] = df_sub['return'].apply(lambda x: np.log(1+x))
            df_sub['return__'] = df_sub['return_'].expanding().sum()
            zmax = df_sub['return__'].max()
            zmin = df_sub['return__'].min()
            crma = np.log(1+zmax) - np.log(1+zmin)
            save_dict[date] = crma
        except:
            save_dict[date] = np.nan
    save_df = pd.DataFrame.from_dict(save_dict,orient='index').reset_index().rename(columns={'index':'day',0:'crma'})
    save_df['code'] = code
    return save_df
```

```python
def applyParallel(dfGrouped, func):
    retLst = Parallel(n_jobs=multiprocessing.cpu_count())(delayed(func)(group) for name, group in dfGrouped)
    return pd.concat(retLst,sort=False)

crma_tmp0 = applyParallel(stock_trade_data_tmp5.groupby('code'), compute_crma)
```

```python
def regression_cross_resvol(df):
    df = df.copy()
    date = df['day'].tolist()[0]
    sub_df = df[['Resvol','Beta','Size','code']].dropna(how='any',axis=0)
    
    y = np.array(sub_df['Resvol'])
    X = np.array(sub_df[['Size','Beta']])
    
    try:
        res = sm.OLS(y,X).fit().resid
    except:
        res = np.nan
    
    sub_df['ResvolO'] = res
    sub_df = sub_df[['ResvolO','code']]
    
    df = df.merge(sub_df,on=['code'],how='left')
        
    return df

def regression_cross_liquidity(df):
    df = df.copy()
    date = df['day'].tolist()[0]
    sub_df = df[['Liquidity','Size','code']].dropna(how='any',axis=0)
    
    y = sub_df['Liquidity']
    X = sub_df['Size']
    
    try:
        res = sm.OLS(y,X).fit().resid
    except:
        res = np.nan
  
    sub_df['LiquidityO'] = res
    sub_df = sub_df[['LiquidityO','code']]
    
    df = df.merge(sub_df,on=['code'],how='left')
        
    return df

def applyParallel(dfGrouped, func):
    retLst = Parallel(n_jobs=multiprocessing.cpu_count())(delayed(func)(group) for name, group in dfGrouped)
    return pd.concat(retLst,sort=False)

resvol_tmp0 = applyParallel(style_factor_df_v3.groupby('day'), regression_cross_resvol)
liquidity_tmp0 = applyParallel(resvol_tmp0.groupby('day'), regression_cross_liquidity)
```

```python
def winsorize(df):
    df = df.copy()
    factors = set(df.columns).difference(['day', 'code'])
    for factor in factors:
        # 某个时间截面上全是NaN
        if df[df[factor].isnull()==True].shape[0] == len(df):
            continue
        median = np.median(df[factor])
        MAD = np.median(np.abs(df[factor] - median))
        df[factor][df[factor]>median+5*MAD] = median+5*MAD  # 盖帽处理偏离中位数5倍以上的数据
        df[factor][df[factor]<median-5*MAD] = median-5*MAD
    return df

def standarlize(df):
    df = df.copy()
    for col in style_factor_list_o:
        if df[df[col].isnull()==True].shape[0] == len(df):
            continue
#         df_sub = df[['code',col]].dropna(how='any',axis=0)
        col_mean = np.mean(df[col])
        col_std = np.std(df[col])
        df[col] = df[col].apply(lambda x: (x-col_mean)/col_std)
#         df_sub = df_sub.rename(columns={col:col+'_afterS'})
#         df = df.merge(df_sub,on=['code'],how='left')
        
    return df

Style_factor_df_o_w = applyParallel(Style_factor_df_o.groupby('day'), winsorize)
Style_factor_df_o_w_s = applyParallel(Style_factor_df_o_w.groupby('day'), standarlize)
```

```python
'''
1.粗略纯因子收益 - 未加国家因子与行业约束，未用加权最小二乘
'''
```

```python
def factor_return(df):
    df = df.copy()
    day = df['day'].tolist()[0]
    X = df[style+industry]
    y = df[['return']]
    return sm.OLS(y,X).fit().params.to_frame().rename(columns={0:day}).T

def applyParallel(dfGrouped, func):
    retLst = Parallel(n_jobs=multiprocessing.cpu_count())(delayed(func)(group) for name, group in dfGrouped)
    return pd.concat(retLst,sort=False)

StyleIndustry_factor_return = applyParallel(StyleIndustry_trade_df_M_dna.groupby('day'),factor_return)
```

```python
def compute_IC(df):
    df = df.copy()
    day = df['day'].tolist()[0]
    return df[style+industry+['return']].corr()['return'].to_frame().rename(columns={'return':day}).T

def applyParallel(dfGrouped, func):
    retLst = Parallel(n_jobs=multiprocessing.cpu_count())(delayed(func)(group) for name, group in dfGrouped)
    return pd.concat(retLst,sort=False)

StyleIndustry_factor_IC = applyParallel(StyleIndustry_trade_df_M_dna.groupby('day'),compute_IC)
```

```python
'''
加权最小二乘法求解每期纯因子收益
权重：用市值平方根代替特质收益率方差的倒数（投资组合管理 P65）
特质收益协方差矩阵：对角阵，对角线为市值平方根
'''
```

```python
def wls_compute_pure_factor(df):
    
    day = df['day'].tolist()[0]
    
    df = df.copy()
    X = df[style+industry]
    y = df['return']
    weight = [1/np.exp(i)**0.5 for i in df['Size'].tolist()]
    
    df_ = sm.WLS(y, X, weight).fit().params.to_frame().rename(columns={0:day}).T
    
    return df_

def applyParallel(dfGrouped, func):
    retLst = Parallel(n_jobs=multiprocessing.cpu_count())(delayed(func)(group) for name, group in dfGrouped)
    return pd.concat(retLst,sort=False)

StyleIndustry_factor_return_wls = applyParallel(StyleIndustry_trade_df_M_dna.groupby('day'),wls_compute_pure_factor)
```

```python
# OLS与WLS纯因子收益对比
save_dct = {}
for factor in style+industry:
    s1 = StyleIndustry_factor_return[factor]
    s2 = StyleIndustry_factor_return_wls[factor]
    pvalue = stats.ttest_rel(s1,s2)[1]
    save_dct[factor] = [pvalue]
t_test_wls_ols = pd.DataFrame(save_dct).T.rename(columns={0:'pair T-test Pvalue'}).sort_values('pair T-test Pvalue')
```

```python
# WLS纯因子收益净值序列
StyleIndustry_factor_equity = StyleIndustry_factor_return_wls[style].copy()
# 剥离日期
day_list = StyleIndustry_factor_equity.index.tolist()
StyleIndustry_factor_equity.index = range(1,len(StyleIndustry_factor_equity)+1)
# 添加初始净值，注意要用loc新增0
StyleIndustry_factor_equity.loc[0,:] = [0 for i in range(len(style))]
StyleIndustry_factor_equity = StyleIndustry_factor_equity.sort_index()

StyleIndustry_factor_equity_result = StyleIndustry_factor_equity.apply(lambda x: 1+x).cumprod()

plt.figure(figsize=(12, 5))
for factor in style:
    equity = StyleIndustry_factor_equity_result[factor]
    plt.plot(equity,label=factor)
#     plt.axhline(tol_df_used.std()['HK_daily_error'], color='r', ls='dashed', lw=1.5,
#                 label='HK_tol_error')

plt.xlabel('day_cnt')
plt.ylabel('pure factor equity')
plt.legend(loc=0)
plt.show()
print('加权最小二乘求解得到的纯因子组合净值序列——每期因子组合对当期该因子暴露为1，其余因子暴露为0')
```
