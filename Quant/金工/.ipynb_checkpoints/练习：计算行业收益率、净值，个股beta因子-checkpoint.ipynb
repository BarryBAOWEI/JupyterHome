```python
## 未处理原始数据
trade_data_daily = pd.read_csv('C:/Users/jxjsj/Desktop/trade_data_daily.csv')
```

```python
## 交易状态添加 1 停牌， 2 未上市， 3 正常
trade_data_daily['type'] = trade_data_daily['volume'].apply(lambda x: 1 if x == 0 else x)
trade_data_daily['type'] = trade_data_daily['volume'].apply(lambda x: 2 if pd.isnull(x) else 3)
```

```python
# 获取行业数据 - 简化了，假设行业样本时间内没发生变化
industry=jq.get_industry(security=SH, date="2019-06-15")
```

```python
# 行业数据合并入行情
trade_data_daily_industry = pd.DataFrame()
for code in SH:
    tmp = trade_data_daily[trade_data_daily['code']==code]
    try:
        tmp['industry'] = industry[code]['sw_l1']['industry_name']
    except:
        tmp['industry'] = 'Other'
    trade_data_daily_industry = pd.concat([trade_data_daily_industry,tmp],axis=0)
```

```python
trade_data_daily_industry_ = trade_data_daily_industry[(trade_data_daily_industry.industry != 'Other') & (trade_data_daily_industry['type'] != 2)]
```

```python
def industry_return(df):
    tmp = df.pivot_table(index='date',columns='code',values='close').pct_change().mean(axis=1).to_frame('return')
    return tmp
industry_return_df = trade_data_daily_industry_.groupby('industry').apply(industry_return).reset_index().set_index('date')
```

```python
def compute_equity(df):
    df_ = df.copy()
    df_['equity'] = df_[['return']].fillna(0).apply(lambda x: np.log(1+x)).expanding().sum().apply(lambda x: np.exp(x)).values
    return df_
industry_return_df_equity = industry_return_df.groupby('industry').apply(compute_equity)
industry_return_df_equity.index = industry_return_df_equity.index.droplevel(0)
```

```python
## 只留下已上市
trade_data_daily_listings = trade_data_daily[trade_data_daily['type'] != 2]
```

```python
## 单独计算收益率以及平均收益率
pivot_price = trade_data_daily_listings.pivot_table(index='date',columns='code',values='close')
pivot_price_ = pivot_price.pct_change()
pivot_price_['MKT'] = pivot_price_.mean(axis=1)
price_ = pivot_price_.unstack()

price_df = price_.reset_index().rename(columns={'o':'return'})
MKT_df = price_df[price_df['code']=='MKT']
```

```python
def merge_price_MKT(df):
    df_ = df.merge(MKT_df,on=['date'],how='left')
    return df_
```

```python
## 普通groupby计算
t1 = datetime.datetime.now()
return_MKT = price_df.groupby('code').apply(merge_price_MKT).rename(columns={'0_x':'return','0_y':'MKT'}).reset_index()[['code','date','return','MKT']]
return_MKT = return_MKT[return_MKT['code']!='MKT']
t2 = datetime.datetime.now()
print(t2 - t1)
```

```python
## 并行groupby计算
t1 = datetime.datetime.now()

def applyParallel(dfGrouped, func):
    retLst = Parallel(n_jobs=multiprocessing.cpu_count())(delayed(func)(group) for name, group in dfGrouped)
    return pd.concat(retLst)

return_MKT = applyParallel(price_df.groupby('code'), merge_price_MKT).rename(columns={'0_x':'return','0_y':'MKT','code_x':'code','code_y':'tmp'}).reset_index()[['code','date','return','MKT']]
return_MKT = return_MKT[return_MKT['code']!='MKT'][['code','date','return','MKT']]
t2 = datetime.datetime.now()
print(t2 - t1)
```

```python
def betacompute(df_from_one_stock):
    df = df_from_one_stock.copy()
    df_dna = df.dropna(axis=0,how='any')
    beta_list = []
    df['beta_'] = np.nan
    for cnt in range(63,len(df_dna)+1):
        df_sub = df_dna.iloc[cnt-63:cnt,]
        y = df_sub['return']
        X = df_sub['MKT']
        X = sm.add_constant(X)
        beta = sm.OLS(y,X).fit().params[1]
        df.iloc[cnt,-1] = beta
        # 测试用
        if cnt == 66:
            break
    return df
```

```python
## 普通groupby计算
t1 = datetime.datetime.now()
beta_df = return_MKT.groupby('code').apply(betacompute)
beta_df.index = beta_df.index.droplevel(0)
t2 = datetime.datetime.now()
print(t2 - t1)
```

```python
## 并行groupby计算
t1 = datetime.datetime.now()

def applyParallel(dfGrouped, func):
    retLst = Parallel(n_jobs=multiprocessing.cpu_count())(delayed(func)(group) for name, group in dfGrouped)
    return pd.concat(retLst)

beta_df = applyParallel(return_MKT.groupby('code'), betacompute)

t2 = datetime.datetime.now()
print(t2 - t1)
```
