```python
def cat_train_valid_test(row):
    if pd.notnull(row['time']):
        td = pd.to_datetime(row['time'][0,10], format='%Y-%m-%d')
        if td <= '2014-12-03':
            return 'train'
        if td >= '2014-12-04':
            return 'valid'
```

```python
train_user['dset'] = train_user.apply(cat_train_valid_test, axis = 1)
```
