```python
# 单个excel输出
def get_all_data(code_lst, year_lst, outputfile):
    """
    parameter statement
    code_lst   : list of stock code, ['000001.XSHE','600519.XSHG',...]
    year_lst   : list of year, ['2017','2016',...]
    """
    for code in code_lst:
        print(code)
        df_code = pd.DataFrame(columns = ['year','non_performing_loan_provision_coverage','Nonperforming_loan_rate','roe','roa','A_E','NL_A','deposit_loan_ratio'])
        cnt = 0
        code_name = jq.get_security_info(code).display_name
        
        q_bank_indicator = jq.query(jq.bank_indicator).filter(jq.bank_indicator.code == code)
        q_indicator = jq.query(jq.indicator).filter(jq.indicator.code == code)
        q_balance = jq.query(jq.balance).filter(jq.balance.code == code)
        
        for year in year_lst:
            print(year)
            
            df_bank_indicator = jq.get_fundamentals(q_bank_indicator, date=None, statDate=year)
            df_indicator = jq.get_fundamentals(q_indicator, date=None, statDate=year)
            df_balance = jq.get_fundamentals(q_balance, date=None, statDate=year)
            
            try:
                non_performing_loan_provision_coverage = float(df_bank_indicator['non_performing_loan_provision_coverage'][0])
                Nonperforming_loan_rate = float(df_bank_indicator['Nonperforming_loan_rate'][0])
                
                roe = float(df_indicator['roe'][0])
                roa = float(df_indicator['roa'][0])
                A_E = roe/roa
                
                total_loan = float(df_bank_indicator['total_loan'][0])
                bad_debts_reserve = float(df_bank_indicator['bad_debts_reserve'][0])
                A = float(df_balance['total_assets'][0])
                NL_A = (total_loan-bad_debts_reserve)/A
                
                deposit_loan_ratio = float(df_bank_indicator['deposit_loan_ratio'][0])
                
            except:
                continue
        
            df_code.loc[cnt] = [year,non_performing_loan_provision_coverage,Nonperforming_loan_rate,roe,roa,A_E,NL_A,deposit_loan_ratio]
            cnt = cnt + 1
            
        out_path = outputfile + code_name + '.xlsx'
        df_code.to_excel(out_path, sheet_name = code_name, index=False)
```

```python
# 输出到一个excel分sheet
def get_all_data_only(code_lst, year_lst, outputfile):
    """
    parameter statement
    code_lst   : list of stock code, ['000001.XSHE','600519.XSHG',...]
    year_lst   : list of year, ['2017','2016',...]
    """
    dict_all = {}
    
    for code in code_lst:
        print(code)
        df_code = pd.DataFrame(columns = ['year','non_performing_loan_provision_coverage','Nonperforming_loan_rate','roe','roa','A_E','NL_A','deposit_loan_ratio'])
        cnt = 0
        code_name = jq.get_security_info(code).display_name
        
        q_bank_indicator = jq.query(jq.bank_indicator).filter(jq.bank_indicator.code == code)
        q_indicator = jq.query(jq.indicator).filter(jq.indicator.code == code)
        q_balance = jq.query(jq.balance).filter(jq.balance.code == code)
        
        for year in year_lst:
            print(year)
            
            df_bank_indicator = jq.get_fundamentals(q_bank_indicator, date=None, statDate=year)
            df_indicator = jq.get_fundamentals(q_indicator, date=None, statDate=year)
            df_balance = jq.get_fundamentals(q_balance, date=None, statDate=year)
            
            try:
                non_performing_loan_provision_coverage = float(df_bank_indicator['non_performing_loan_provision_coverage'][0])
                Nonperforming_loan_rate = float(df_bank_indicator['Nonperforming_loan_rate'][0])
                
                roe = float(df_indicator['roe'][0])
                roa = float(df_indicator['roa'][0])
                A_E = roe/roa
                
                total_loan = float(df_bank_indicator['total_loan'][0])
                bad_debts_reserve = float(df_bank_indicator['bad_debts_reserve'][0])
                A = float(df_balance['total_assets'][0])
                NL_A = (total_loan-bad_debts_reserve)/A
                
                deposit_loan_ratio = float(df_bank_indicator['deposit_loan_ratio'][0])
                
            except:
                continue
        
            df_code.loc[cnt] = [year,non_performing_loan_provision_coverage,Nonperforming_loan_rate,roe,roa,A_E,NL_A,deposit_loan_ratio]
            cnt = cnt + 1

        dict_all[code_name] = df_code
        
    # 定义一个将字典输出为excel的函数
    def stock_to_excel(outputpath,df_dict):
        """
        outputpath: str  输出路径，例如'c:/XXX/YYYYY/'
        df_dict   : dict stock_inf_get函数得到的结果字典
        """
        outputfile = outputpath + 'stock_inf.xlsx'
        writer = pd.ExcelWriter(outputfile)
        for key, value in df_dict.items():
            value.to_excel(writer,key,index=False)
        writer.save()     
    
    stock_to_excel(outputfile, dict_all)
```
