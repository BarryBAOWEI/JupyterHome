```python
import tushare as ts
import pandas as pd
import datetime
```

```python
def trans_d_to_m(df):
    """
    df: DataFrame 必须包含'date'列，格式为yyyy-mm-dd
    """
    dff = df
    dff['y-m'] = dff['date'].apply(lambda x : str(x)[0:7])
    df_temp = pd.DataFrame()
    df_temp['date'] = dff['date'].groupby(dff['y-m']).max()
    dff = pd.merge(left = df_temp, right = dff, how = 'inner', left_on = 'date', right_on = 'date')
    # df['return'] = df_m_return['close'].pct_change()
    return dff
```

```python
################### test ###################
df = ts.get_k_data('600580',start = '2000-01-01',end = '2018-05-10')
print(df)
print(trans_d_to_m(df))
```
