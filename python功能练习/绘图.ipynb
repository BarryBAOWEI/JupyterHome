```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 使用plt绘图 统一绘图 图例添加失败
from sklearn import datasets
iris = datasets.load_iris()
x_temp = pd.DataFrame(iris.data, columns = ['Sepal.L','Sepal.W','Petal.L','Petal.W'])
y_temp = pd.DataFrame(iris.target, columns =['Class'])
y_temp['Class'] = y_temp['Class'].apply(lambda x: {0:'setosa',1:'versicolor',2:'virginica'}[x])

def ColorCode(df):
    df_temp = df
    colors = ['b','g','r','c','m','tan']
    columns_names = df_temp.columns.tolist()
    for columns_name in columns_names:
        features_lst = set(df_temp[columns_name])
        dct = {}
        code = 0
        for feature in features_lst:
            dct[feature] = code
            code += 1
        df_temp[columns_name] = df_temp[columns_name].apply(lambda x: colors[dct[x]])
    return df_temp
y_color = ColorCode(y_temp)['Class'].tolist()

# x:横坐标 y:纵坐标 s:点的尺寸
plt.scatter(x_temp['Sepal.L'], x_temp['Sepal.W'], c = y_color)

# 图例
# plt.legend(labels = ['setosa','versicolor','virginica'], loc='upper right')
 
# 设置图表标题并给坐标轴加上标签
plt.title('iris Data', fontsize=24)
plt.xlabel('Sepal.L', fontsize=14)
plt.ylabel('Sepal.W', fontsize=14)
 
# 设置刻度标记的大小
# plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```

```python
# 将数据集按照y标签拆分 再绘图
from sklearn import datasets
iris = datasets.load_iris()

# 建立DataFrame，合并，修改y的名称
x_temp = pd.DataFrame(iris.data, columns = ['Sepal.L','Sepal.W','Petal.L','Petal.W'])
y_temp = pd.DataFrame(iris.target, columns =['Class'])
data = pd.concat([x_temp, y_temp], axis = 1)
data['Class'] = data['Class'].apply(lambda x: {0:'setosa',1:'versicolor',2:'virginica'}[x])

# 按Class列拆分DF
data_setosa = data[data['Class'].isin(['setosa'])]
data_versicolor = data[data['Class'].isin(['versicolor'])]
data_virginica = data[data['Class'].isin(['virginica'])]

# 每种Class分别绘图
p1 = plt.scatter(data_setosa['Sepal.L'],data_setosa['Sepal.W'], c = 'b')
p2 = plt.scatter(data_versicolor['Sepal.L'],data_versicolor['Sepal.W'], c = 'g')
p3 = plt.scatter(data_virginica['Sepal.L'],data_virginica['Sepal.W'],c = 'r')

# 设置图表标题并给坐标轴加上标签
plt.title('iris Data', fontsize=24)
plt.xlabel('Sepal.L', fontsize=14)
plt.ylabel('Sepal.W', fontsize=14)

plt.legend(handles = [p1,p2,p3], labels = ['setosa','versicolor','virginica'], loc = 'best')
```

```python
from sklearn import datasets
iris = datasets.load_iris()
x_temp = pd.DataFrame(iris.data, columns = ['Sepal.L','Sepal.W','Petal.L','Petal.W'])
y_temp = pd.DataFrame(iris.target, columns =['Class'])
data = pd.concat( [x_temp, y_temp], axis=1 )
data['Class'] = data['Class'].apply(lambda x: {0:'setosa',1:'versicolor',2:'virginica'}[x])

# 散点图矩阵
# Seaborn visualization library
import seaborn as sns
# Create a pair plot colored by continent with a density plot of the # diagonal and format the scatter plots.
# sns.pairplot(data, hue = 'Class', diag_kind = 'kde',
#              plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'},
#              size = 4)

# Plot colored by continent for years 2000-2007
sns.pairplot(data[data['Sepal.L'] >= 4][data['Sepal.L']<=5.4], 
             vars = ['Sepal.W','Petal.L','Petal.W'], 
             hue = 'Class', diag_kind = 'kde', 
             plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'},
             size = 3)
# Title 
plt.suptitle('Pair Plot of iris Data with Sepal.L>=4', 
             size = 18)
```
