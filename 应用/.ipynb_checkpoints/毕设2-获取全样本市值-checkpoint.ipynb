```python
import pandas as pd
import numpy as np
import jqdatasdk as jq
import os
```

```python
jq.auth('13918852005','960312Lsc')
# jq.auth('15821912507','912507')
```

```python
stock_download_lst = pd.read_csv('E:/Stock_Data/stock_code.csv')
stock_download_lst = list(stock_download_lst.iloc[:,0])
```

```python
trade_days = jq.get_trade_days(start_date='2011-01-01', end_date='2018-12-31')
trade_days_len = len(trade_days)
```

```python
used_stock_list = []
cnt_code = 0
code_list_temp = os.listdir('E:/Stock_Data/stock_trade_data')
code_list = [i[:-4] for i in code_list_temp]

for code in stock_download_lst:
    
    if code in code_list:
        cnt_code += 1
        continue
    
    cnt = 0
    for trade_day in trade_days:
        trade_inf = jq.get_fundamentals(jq.query(
            jq.valuation.day,
            jq.valuation.capitalization,
            jq.valuation.market_cap                        
                                    ).filter(jq.valuation.code==code),trade_day)
        if cnt ==0:
            # 若第一天就无数据，剔除
            if len(trade_inf)==0:
                break
            trade_save_df = trade_inf
        else:
            trade_save_df = pd.concat([trade_save_df,trade_inf],axis=0)               
        cnt += 1       
        # 若数据下载过半，有效数据小于一般，剔除
        if cnt>trade_days_len/2 and len(trade_save_df)<trade_days_len/2:
            break
        print('No.'+str(cnt_code+1)+" Complete:%.4f%%" % (cnt/trade_days_len*100),end="\r")
    
    # 保存
    cnt_code += 1
    used_stock_list.append(code)
    trade_save_df.to_csv('E:/Stock_Data/stock_trade_data/'+code+'.csv',index=False)
    
pd.DataFrame(used_stock_list).to_csv('E:/Stock_Data/stock_trade_data/trade_inf_used_stock.csv',index=False)
```
