```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plot_df = pd.read_csv('E:/Stock_Data/calander effect test sum/4of5FactorThu_plot.csv',index_col=0)
```

```python
X = list(range(1,9))
```

```python
fig = plt.figure(1)
for i in range(4):
    name = plot_df.index[i]
    if name == 'B_M':
        name = 'B/M'
    plot_num = 221+i
    plt.subplot(plot_num)
    y = plot_df.iloc[i,:]
#     plt.plot(X, y, 'ks','r-', lw=2)
    plt.title('Bin - '+name, fontsize=14)
    plt.xlabel('Small   â†’   Big', fontsize=14)
    plt.ylabel('E[R] of Thu', fontsize=14)
    plt.plot(X, y)
fig.tight_layout()
plt.show()
```
