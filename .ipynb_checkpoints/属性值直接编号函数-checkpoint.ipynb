```python
def FeatureCode(df):
    columns_names = df.columns.tolist()
    for columns_name in columns_names:
        features_lst = set(df[columns_name])
        dct = {}
        code = 1
        for feature in features_lst:
            dct[feature] = code
            code += 1
        df[columns_name] = df[columns_name].apply(lambda x: dct[x])
```
